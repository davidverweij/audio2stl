```r
source("audio2stl.r", local = knitr::knit_global())

message("script loaded succesfully")
```

As described in Christopher Carignan's [README](README.md) and [audio2stl.r](audio2stl.r), the `audio2stl()` method takes 6 parameters:

required:
- inputfile: filepath string associated with WAV audio file

optional:
- outputfile: filepath sring associated with STL file to save
- sampfreq: re-sampling frequency
- axisnorm: keep the x- and y-axes scaling (FALSE) or normalize so that they have the same dimensions (TRUE)
- preemph: value for pre-emphasis of audio
- window: option for windowing audio using a variety of Matlab/Octave compatible filters found in the "signal" package. E.g. 'bartlett', 'blackman', 'hamming', 'hanning', 'triang'

```r
files <- list.files(path="files/", pattern="*.wav", full.names=FALSE, recursive=FALSE)
lapply(files, function(x) {
    input = paste("files/", x,sep="")
    message(input)
    out = paste("output/converted_", x,sep="")
    audio2stl(input, out)
    message(out)
    # t <- read.table(x, header=TRUE) # load file
    # apply function
    # out <- function(t)
    # write to file
    # write.table(out, "path/to/output", sep="\t", quote=FALSE, row.names=FALSE, col.names=TRUE)
})


# to save to STL use:
# audio2stl("phonetics.wav", "result.stl")

# David: It is unclear to me why this screen is black. When saving the STL it does show a proper model
```
